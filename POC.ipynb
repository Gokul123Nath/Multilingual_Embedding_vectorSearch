```python
co = cohere.Client('h7BI3irdxYktGqLEpKjfUwOFerutik4FLfXuooTI')
texts = ['こんにちは、元気ですか','一番高い山はエベレストです','英語は何文字ありますか','英語の文字は26文字あります','私の車は一番速いです']
text_eng = ['hi,how are you', 'The tallest mountain is evrest', 'how many letters are there in english','There are 26 letters in english','My car is fastest']
response = co.embed(texts=texts, model='multilingual-22-12')
embeddings = response.embeddings
print(embeddings)
```

```python
from annoy import AnnoyIndex
import numpy as np
search_index = AnnoyIndex(np.array(embeddings).shape[1],'angular')
```

```python
for i in range(len(embeddings)):
    search_index.add_item(i, embeddings[i])
```

```python
search_index.build(10) # 10 trees
search_index.save('test.ann')
```

```python
query = 'ஆண்டின் கார்'
query_embed = co.embed(texts=[query], model='multilingual-22-12', input_type='search_query').embeddings
print(query_embed[0])
```

```python
similar_item_ids = search_index.get_nns_by_vector(query_embed[0],2,include_distances=True)
```
